| Sheet | Description | Key Columns |
|-------|-------------|-------------|
| Shortage | Contains materials that require replenishment and the 7-day shortage quantity. | Material, 7Day Shortage, Need Date |
| Stock | Contains current inventory by warehouse and storage location. | Material, Warehouse, Sloc, Qty |
| Open_PO | Contains supplier information, including vendor available stock, purchase order quantity, and ETA. | Material, Vendor Available Qty, PO Qty, ETA |
| SLOC_PRIORITY | Defines the priority of storage locations for stock transfer. | Sloc, Priority |
