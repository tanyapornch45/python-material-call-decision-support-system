## Business Logic

1. Read Shortage Report  
2. Read Stock Report  
3. Calculate Total Available Quantity  
4. Calculate Stock Balance  

**If Stock Balance ≥ 0**
- Generate Transfer Plan
- Decision: Transfer Stock

**If Stock Balance < 0**
- Calculate Call Quantity
- Check Vendor Available Quantity

**If Vendor Available Quantity ≥ Call Quantity**
- Decision: Call Material

**If Vendor Available Quantity < Call Quantity**
- Check ETA

**If ETA is available and ETA ≤ Need Date**
- Decision: Wait for Material Arrival

**Otherwise**
- Decision: Please Pull In ETA
