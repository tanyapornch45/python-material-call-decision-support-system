# Python-based Material Call Decision Support System

# Project Overview

## Project Name

Python-based Material Call Decision Support System for Semiconductor Manufacturing

## Background

In semiconductor manufacturing, material shortages can interrupt production and delay customer deliveries. Material planners need to quickly determine whether shortages can be resolved by transferring available stock from another warehouse or by waiting for incoming purchase orders.

This decision is often made manually by checking multiple Excel reports from SAP, which is time-consuming and prone to human error.

## Project Objective

This project aims to automate the material call decision process using Python and Pandas.

The system analyzes shortage reports, available stock, vendor available stock, ETA information, and SLOC priorities to generate material replenishment decisions.

## Expected Benefits

- Reduce manual material checking time.
- Improve the consistency of replenishment decisions.
- Support data-driven material planning.
- Demonstrate the practical application of Python and Pandas in semiconductor supply chain management.
