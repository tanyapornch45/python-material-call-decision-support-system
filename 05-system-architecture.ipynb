# System Architecture

The proposed system integrates multiple SAP reports into a single decision support workflow.

The system reads shortage reports, stock reports, open purchase orders, and SLOC priorities, then processes the data using Python and Pandas to generate replenishment recommendations.

### System Components

Input Reports
- Shortage
- Stock
- Open PO
- SLOC Priority

Processing
- Read Excel files
- Calculate available quantity
- Evaluate vendor availability
- Check ETA
- Generate decision

Output
- Transfer Plan
- CALL Material
- Wait for Material Arrival
- PLS PULL IN ETA
