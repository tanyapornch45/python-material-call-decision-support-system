# Problem Statement

In semiconductor manufacturing, material shortages occur frequently due to demand fluctuations, supply delays, and inventory constraints. If shortages are not handled promptly, they may interrupt production and delay customer deliveries.

Currently, material planners manually review multiple SAP reports, including shortage reports, stock reports, open purchase orders, and warehouse information. They need to compare inventory levels, check incoming purchase orders, and determine the most appropriate replenishment action.

This manual process requires switching between several Excel files and performing repetitive checks, which is time-consuming and may lead to inconsistent decisions or human errors.

To improve this process, a Python-based decision support system is proposed. The system integrates multiple input reports, evaluates stock availability and open purchase orders, and automatically generates replenishment recommendations. This approach helps reduce manual effort, improve decision consistency, and support faster material planning.

Based on this problem, the proposed system evaluates material shortages by checking available stock, warehouse priorities, and open purchase orders before generating replenishment recommendations.
